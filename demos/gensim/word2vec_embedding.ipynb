```python
import gensim.downloader as api
from gensim.models import Word2Vec
dataset = api.load('text8')
```

```python
model = Word2Vec(dataset)
```

```python
model.save("text8-word2vec.bin")
```

```python
from gensim.models import KeyedVectors
model = KeyedVectors.load("text8-word2vec.bin")
word_vectors = model.wv
```

```python
word_vectors.vocab
```

```python
word_vectors.most_similar('king')
```

```python
word_vectors.most_similar(positive = ['france' , 'berlin'] , negative = ['paris'])[0:5]
```

```python
word_vectors.similarity('hindu', 'allah')
```

```python
word_vectors.similar_by_word('muslims')
```

```python
word_vectors["song"]
```

```python
word_vectors.word_vec('hindu', use_norm =True)
```
